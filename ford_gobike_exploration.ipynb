**Trip duartion has an extreme values, this leads to unclear box plots or violin plots, let's cut outliers**

```python
def GetMinMaxValues(data):
    IQR = data.quantile(0.95)-data.quantile(0)
    minValue = (data.quantile(0))-(1.5*IQR)
    maxValue = (data.quantile(.95))+(1.5*IQR)
    return minValue,maxValue

durationMinValue,durationMaxValue = GetMinMaxValues(df['duration_sec'])
df_without_outliers = df[(df['duration_sec'] > durationMinValue) & (df['duration_sec'] < durationMaxValue)] 
```

```python
plt.figure(figsize = [8,8])
sns.boxplot(y = 'duration_sec',x = 'user_type',data = df_without_outliers,palette = "Set3")
plt.title("Trip duration vs User type")
plt.xlabel('User type')
plt.ylabel('Trip Duration (sec)')
pd.DataFrame(df_without_outliers.groupby(['user_type'])['duration_sec'].describe().loc[:,['mean','std']])
```

Customer user type has a larger mean trip duration than subsriber

```python
plt.figure(figsize = [8,8])
sns.boxplot(y = 'duration_sec',x = 'bike_share_for_all_trip',data = df_without_outliers,palette = "Set2")
plt.title("Trip duration vs bike share for all trip")
plt.xlabel('Bike share for all trip')
plt.ylabel('Duration (sec)')
pd.DataFrame(df_without_outliers.groupby(['bike_share_for_all_trip'])['duration_sec'].describe().loc[:,['mean','std']])
```

The plot shows that bike share for all trip has no significant effect on trip duartion

```python
plt.figure(figsize = [8,8])
sns.boxplot(y = 'duration_sec',x = 'member_gender',data = df_without_outliers,palette = "Set2")
plt.title("Trip duration vs Member gender")
plt.ylabel('Duration (sec)')
pd.DataFrame(df_without_outliers.groupby(['member_gender'])['duration_sec'].describe().loc[:,['mean','std']])
```

The plot shows that higher percentage of females have longer trips than males or other
