## Investigation Overview

The main goal of this investigation is to find the characterstics of bike users that could affect the trip duration. I will focus on these attributes: Member age and user type.

## Dataset Overview

There are 183412 rides in the dataset with 16 features (duration, start time, end time, start and end stations info :id, name latitude and longitude, bike id, user type, member birth year, member gender and bike share for all trip).
Some variables are numerical and other are nominal categorical variables.
Time Frame of the data: February 2019

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv("FordGoBike.csv")
```

## Trip duration distribution

Trip duration has a wide range of values, starting from 1 min and ending at about 1400 min, the distribution has a long tail with a peak at 10 min.

```python
plt.figure(figsize = [13,9])
sns.histplot(df['duration_sec'],log_scale = True, bins = 50)
plt.title('Trip duration distribution')
plt.xlabel('Trip duration')
plt.ylabel('Number of trips');
```

## Member age distribution

Most members are between 20 and 40 years old

```python
plt.figure(figsize = [13,9])
sns.histplot(df['age'], bins = 50)
plt.title('Age distribution')
plt.xlabel('Age')
plt.ylabel('Number of trips');
```

## Trip duration and member age relatioship

Long trips were done by younger users

```python
plt.figure(figsize = [10,7])
sns.scatterplot(data=df, x='age', y='duration_sec',color = 'green');
plt.title('Age Vs Trip duration')
plt.xlabel('Age (year)')
plt.ylabel('Trip duration (sec)');
```

## User type counts

Most trips done by subscribers, they are almost 8 times than customers.

```python
plt.figure(figsize = [10,7])
sns.countplot(x = 'user_type', data = df)
plt.title('User type Counts')
plt.xlabel('User type')
plt.ylabel('Number of trips');
```

## Trip duration and User type relationship

Customer user type has a larger mean trip duration than subsriber

```python
plt.figure(figsize = [8,8])
sns.boxplot(y = 'duration_sec',x = 'user_type',data = df,palette = "Set3")
plt.title("Trip duration vs User type")
plt.xlabel('User type')
plt.ylabel('Trip Duration (sec)')
plt.ylim(0,3000)
```
